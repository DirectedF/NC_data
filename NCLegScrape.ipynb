```python
import requests
from bs4 import BeautifulSoup

url = "https://www.ncleg.gov/Members/MemberTable/S"

response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

# parent_div = soup.find("div", {"class": "row ng-scope"})
# memberTable

table = soup.find("table", attrs={"class": "display responsive table table-striped table-bordered"})

members = []

for row in table.find_all("tr")[1:]:
    cols = row.find_all("td")
    member = {
        "party": cols[0].get_text(),
        "district": cols[1].get_text(),
        "districtmap": cols[2].get_text(),
        "memberLast": cols[3].get_text(),
        "memberFull": cols[4].get_text(),
        "counties represented": cols[5].get_text() if cols[5].get_text() != "" else None
    }
    members.append(member)

for member in members:
    print(member["party"],member["counties represented"])
```
